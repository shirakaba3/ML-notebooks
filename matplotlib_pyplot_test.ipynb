import 

```python
import matplotlib.pyplot as plt # お約束のニックネーム

import numpy as np
```

```python
y = np.random.rand(100)
x = np.linspace(1, 5, 100)
```

## 最もシンプルなグラフ
まずは何も考えずにグラフをplotしてみる

```python
plt.plot(y)
```

```python
# グラフにラベルを付けてみる

plt.plot(x,y, label="random data")

plt.xlabel("time series")
plt.ylabel("power")

plt.title("sample graph")

plt.legend()
```

### 複数のグラフを表示したい

```python
x = np.linspace(0, 5*np.pi, 100)

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label="sin curve")
plt.plot(x, y_cos, label="cos curve")

plt.legend()
```

## scatter plot

```python
x = np.random.randn(50)
y = np.random.randn(50)
plt.scatter(x,y)
```

```python
x1 = np.random.randn(50)
y1 = np.random.randn(50)
plt.scatter(x1,y1, label="x1")

x2 = np.random.randn(50)
y2 = np.random.randn(50)
plt.scatter(x2,y2, label="x2")
```

## 円グラフ

```python
x = np.array([10, 5,3, 12,6,4])

cate = ["A","B","C", "D", "E", "F"]
plt.pie(x, labels=cate)
```

##  棒グラフ

```python
plt.bar(cate ,x)
```
