## snsで使える便利な機能

### データセットがsnsで読み込める

```python
# snsで使えるデータ一覧

sns.get_dataset_names()
```

```python
iris = sns.load_dataset("iris")

iris.head()
```

### データの散布図を全ての変数の組み合わせ分表示


```python
sns.pairplot(iris)
```

```python
# hueに指定したcolumnを基準にデータを色分け

sns.pairplot(iris, hue="species")
```

```python
# iris以外のデータについても可視化してみよう

data = sns.load_dataset("attention")

data.head()
```

```python
sns.pairplot(data, hue="attention")
```

### 特定の特徴の組み合わせの散布図

```python
sns.jointplot('sepal_width', 'petal_length', data=iris)
```

```python
# 表示を六角形でかっこよく変更
sns.jointplot('sepal_width', 'petal_length', data=iris, kind="hex")
```

```python
# 等高線で密度を表示
sns.jointplot('petal_width', 'petal_length',data=iris, kind="kde")
```

```python
# 線形回帰もできる
sns.jointplot('petal_width', 'petal_length', data=iris, kind="reg")
```

### 相関関係の可視化と階層型クラスタリングでデンドログラムを作成

```python
sns.clustermap(iris.corr())
```
